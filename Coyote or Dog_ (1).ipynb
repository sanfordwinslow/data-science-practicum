```python
categories = []
length =0
#create set of filenames and
filenames = []
for filepath in files_in_zip:
    filename = filepath.split('/')[1]
    category = filepath.split('/')[1].split('.')[0]
    #print(filename)
    #add each file name to it
    filenames.append(filename)
    length +=1
    if category == 'dog':
        categories.append(1)
    else:
        categories.append(0)
print(length)        
# Adding filenames to the dataframe throws errors most likely
# because there is different amount of file names that categorries?
# I was using filename for dictionary, but got error, and replaced with filepath
df = pd.DataFrame({
    'filenames': filenames,
    'category': categories
})


print(df)
```
